**1**. (25 points)

The code below gives five "documents" with titles in `titles` and text in `contents`. 

- Convert each text into "words" by converting to lower case, removing punctuation and splitting on whitespace
- Make a list of all unique "words" in any of the texts
- Create an pandas DataFrame whose rows are words, columns are titles, and values are counts of the word in the document
- Add a column `total` that counts the total number of occurrences for each word across all documents
- Show the rows for the 5 most commonly used words

````python
import sklearn
from sklearn.datasets import fetch_20newsgroups
twenty = fetch_20newsgroups(subset='train')
target_names = twenty['target_names']
titles = [target_names[i] for i in twenty['target'][2:7]]
contents = twenty['data'][2:7]
````

### Q1 Solution 

````python
#1 Convert each text into "words" by converting to lower case, removing punctuation and splitting on whitespace
import string 
words=[text.lower().translate(str.maketrans('','',string.punctuation)).split() for text in contents]
words
````

````python
#2 Make a list of all unique "words" in any of the texts 
vocals=set([word for item in words for word in item])
vocals
````

````python
#3 Create an pandas DataFrame whose rows are words, columns are titles, and values are counts of the word in the document
import numpy as np
import pandas as pd
table=np.zeros((len(vocals),len(titles)),dtype=int)
for i, vocal in enumerate(vocals):
    for j, word in enumerate(words):
        table[i,j]=word.count(vocal)
data=pd.DataFrame(data=table,columns=titles,index=vocals)
data.head()
````

````python
#4 Add a column total that counts the total number of occurrences for each word across all documents
total=[sum(i) for i in table]
data2=pd.DataFrame(total,columns=['Total'],index=vocals)
data3=pd.merge(left=data,right=data2,left_index=True,right_index=True)
data3.head()
````

````python
# Show the rows for the 5 most commonly used words
ind=data3.sort_values(by='Total',ascending=False)[0:5].index
data3.loc[ind]
````

For example, if the offset is 3, we replace `a` with `d`, `b` with `e` etc. The cipher wraps around so we replace `y` with `b`, `z` with `c` and so on. ### Q2 Solution

````python
#1 Write a function encode that takes as arguments a string and an integer offset and returns the encoded cipher
def encode(ss,k):
    '''
    This is an encode function that takes as arguments a string 'ss' and an integer offset 'k' and returns the encoded cipher
    '''
    k=k%26
    a=string.ascii_lowercase 
    b=string.ascii_uppercase
    origin=string.ascii_letters
    code=a[k:]+a[:k]+b[k:]+b[:k]
    return ss.translate(str.maketrans(origin,code))
````

````python
#2 Write a function decode that takes as arguments a cipher and an integer offset and returns the decoded string

def decode(ss,k):
    '''
    This is a decode function that takes as arguments a cipher 'ss' and an integer offset 'k' and returns the decoded string
    '''
    k=k%26
    a=string.ascii_lowercase 
    b=string.ascii_uppercase
    origin=string.ascii_letters
    code=a[k:]+a[:k]+b[k:]+b[:k]
    return ss.translate(str.maketrans(code,origin))
````

````python
#test encode function, offset=3
cipher=encode(ss,3)
print(cipher)
````

````python
# test decode function, offset k=3
strings=decode(cipher,3)
print(strings)
````

````python
#3 Write a function auto_decode that takes as argument a cipher and uses a statistical method to guess the optimal offset to decode the cipher


def order_text(ss,k): 
    '''
    this function is to find the frequency order of given string ss decoded by offset k(ignore letter case)
    '''
    k=k%26
    a=string.ascii_lowercase 
    s1=ss.lower().translate(str.maketrans('','',string.punctuation+string.whitespace))
    s1=decode(s1,k)
    return pd.DataFrame(data=[s1.count(i) for i in a],index=[a for a in a],columns=['freq1']).sort_values(by='freq1',ascending=False).index


freq = {'a': 0.08167, 'b': 0.01492, 'c': 0.02782, 'd': 0.04253, 'e': 0.12702, 'f': 0.02228,
 'g': 0.02015, 'h': 0.06094, 'i': 0.06966, 'j': 0.00153, 'k': 0.00772, 'l': 0.04025, 'm': 0.02406, 'n': 0.06749,
 'o': 0.07507, 'p': 0.01929, 'q': 0.00095, 'r': 0.05987, 's': 0.06327, 't': 0.09056, 'u': 0.02758, 'v': 0.00978,
 'w': 0.0236, 'x': 0.0015, 'y': 0.01974, 'z': 0.00074}

## this the letter frequency order(from most freqnent to less frequent)
order1=pd.DataFrame(data=pd.Series(data=freq),columns=['freq']).sort_values(by='freq',ascending=False).index



def auto_decode(ss):
    '''
    ss is a cipher and this function will return a best match decode offset k based on the frequency of letter used in words:
    find the offset such that the letter frequency order is largely overlap with given order.
    
    this function return the decoded string
    '''
    test={}
    for k in range(26):
        match=0
        for i in range(26):
            if order1[i]==order_text(ss,k)[i]:# if the decoded strings' letter frequent order is the same as given order, match takes1
                match+=1
        test[k]=match
    k=pd.Series(data=test).idxmax()# the optimal offset is given by an offset that max the match
    
    return decode(ss,k)
````

````python
import random
cipher_r=encode(ss,random.randint(10,20))
print(cipher_r)
````

````python
decoded_string=auto_decode(cipher_r)
print(decoded_string)
````
